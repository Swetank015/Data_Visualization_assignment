**Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
n_samples, n_features = 100, 5
X = np.random.randn(n_samples, n_features)
corr_matrix = np.array([[1, 0.8, 0.5, 0.2, 0.1],
                        [0.8, 1, 0.7, 0.3, 0.2],
                        [0.5, 0.7, 1, 0.4, 0.3],
                        [0.2, 0.3, 0.4, 1, 0.5],
                        [0.1, 0.2, 0.3, 0.5, 1]])
X = np.dot(X, np.linalg.cholesky(corr_matrix))

df = pd.DataFrame(X)

sns.set_theme()
sns.heatmap(df.corr(), annot=True, cmap="viridis")
plt.title("Correlation Matrix")
plt.show()
```

**Q3.Using the sales data, generate a heatmap to visualize the variation in sales across
different months and days.**

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
sales_df = pd.DataFrame(data)

df_pivoted = sales_df.pivot_table(index='Day', columns='Month', values='Sales')

fig = go.Figure(data=go.Heatmap(z=df_pivoted.values, x=df_pivoted.columns, y=df_pivoted.index, colorscale='Viridis'))

fig.update_layout(title='Sales Heatmap', xaxis_title='Month', yaxis_title='Day')

fig.show()
```

**Q5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.**

```python
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK','Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000,5)
}
df = pd.DataFrame(data)
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK','Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000,5)
}
df = pd.DataFrame(data)

colorscale = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df['GDP'],
    y=df['Population'],
    mode='markers',
    marker=dict(
        size=df['Population'] * 2,
        color=colorscale[:len(df['Country'])],
        opacity=0.7
    ),
    text=df['Country'],
))

fig.update_layout(
    title='Bubble Chart: GDP vs Population',
    xaxis_title='GDP',
    yaxis_title='Population',
    hovermode='closest'
)

fig.show()
```

**Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x**

```python
from bokeh.plotting import figure, show
import numpy as np
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

x = np.linspace(0, 10, 1000)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, line_width=3)

show(p)
```

**Q2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the

**Q4. Create a Bokeh histogram to visualize the distribution of the given data.**

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
from bokeh.plotting import figure, show
import numpy as np

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", tools='', background_fill_color="#fafafa")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

p.y_range.start = 0
p.xaxis.axis_label = 'x'
p.yaxis.axis_label = 'Pr(x)'
p.grid.grid_line_color="white"

show(p)
```
